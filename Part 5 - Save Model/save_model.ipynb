```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("houseprices.csv", sep="\t")
df.head()
```

```python
model = linear_model.LinearRegression()
model.fit(df[['area']],df.price)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[5000]])
```

```python
import pickle

with open('model_pickle', 'wb') as f:
    pickle.dump(model, f)
```

```python
with open("model_pickle", 'rb') as f:
    mp = pickle.load(f)
```

```python
mp.predict([[5000]])
```

```python
import joblib

joblib.dump(model, 'model_joblib')
```

```python
mj = joblib.load('model_joblib')

mj.predict([[5000]])
```
